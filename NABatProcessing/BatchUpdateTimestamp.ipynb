### An example of batch updating the date time in a group of wav files

```python
from pathlib import Path
import datetime
from guano import GuanoFile
import shutil
```

```python
source_dname = r"C:\temp\to_rename"

wavs = Path(source_dname).glob('**\*.wav')

out_dname = r"C:\temp\renamed"
out_d = Path(out_dname)
```

```python
for wav in wavs:
    print(wav)
    moved_copy = out_d.joinpath(wav.name.replace('_2015', '_2016'))
    shutil.copy(wav, moved_copy)
    
    g = GuanoFile(moved_copy)
    g['Timestamp'].replace(year = g['Timestamp'].year + 1)
    g.write(make_backup=False)
```
